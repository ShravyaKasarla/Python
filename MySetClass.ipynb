```python
class MySet:
    
    def __init__(self,s):
        self.set_s(s)
            
    
    def add(s, value):
        s |= set((value, ))
        
    
    def clear(s):
        s &= set()

    def copy(s):
        return s
    
    
    def difference(s1, s2):
        return s1-s2
    
    
    def difference_update(s1, s2):
        s1 -= s2
        
        
    def discard(s, value):
        if value in s:
            for i in s:
                if i == value:
                    s -= set((value, ))
                    return None
        return s
    
    
    def intersection(s1, s2):
        return s1 & s2
    
    
    def intersection_update(s1, s2):
        s1 &= s2
        
        
    def isdisjoint(s1, s2):
        if s1 & s2 == set():
            return True
        return False
    
    
    def issubset(s1, s2):
        if s1 <= s2:
            return True
        return False
    
    
    def issuperset(s1, s2):
        if s1 >= s2:
            return True
        return False
    
    
    def pop(s):
        import random
        pop_value = random.choice(list(s))
        for i in s:
                if i == pop_value:
                    s -= set((pop_value, ))
                    break
        return pop_value
    
    
    
    def remove(s, value):
        if value in s:
            for i in s:
                if i == value:
                    s -= set((value, ))
                    return None
        return KeyError
    
    
    
    def symmetric_difference(s1, s2):
        return s1^s2  #(s1-s2)|(s2-s1)
 

    
    def symmetric_difference_update(s1, s2):
        s1 ^= s2
        
        
    def union(s1, s2):
        return s1 | s2


    def update(s1, s2):
        s1 |= s2
```
