```python
class MyDictionary:
    
    #1.clear
    def dict_clear(d):
        '''Remove all items from D and returns None.'''
        n = len(d)
        for i in range(n):
            key = list(d)[-1]
            pop_value = d[key]
            del d[key]
        
        
    #2.copy
    def dict_copy(d):
        '''returns a shallow copy of D'''
        return d
    
    
    #3.fromkeys
    def fromkeys(itr, value = None):
        '''Create a new dictionary with keys from iterable and values set to value.'''
        d = {}
        for i in itr:
            d[i] = value
        return d
    
    
    #4.get
    def get(d, key, default=None):
        '''Return the value for key if key is in the dictionary, else default.'''
        if key in list(d):
            return d[key]
        
        
    #5.items
    def items(d):
        '''returnsa list-like object providing a view on D's items'''
        l=[]
        for i in d:
            l += [(i, d[i])] 
        return l
    
    
    #6.keys
    def keys(d):
        '''a list-like object providing a view on D's keys'''
        return list(d)
    
    
    #7.pop
    def pop(d, key):
        '''removes the specified key and return the corresponding value.
        If the key is not found, return the default if given; otherwise,
    raise a KeyError.'''
        pop_value = d[key]
        del d[key]
        return pop_value
    
    
    #8.popitem
    def popitem(d):
        '''Remove and return a (key, value) pair as a 2-tuple.'''
        key = list(d)[-1]
        pop_value = d[key]
        del d[key]
        return key, pop_value
    
    #9.setdefault
    def setdefault(d, key, default = None):
        '''Insert key with a value of default if key is not in the dictionary.'''
        if key in d:
            return d[key]
        d[key] = default
        
    #10.update
    def update(d, items):
        for i in items:
            d[i] = items[i]
            
    #11.values
    def values(d):
        l =[]
        for i in d:
            l += [d[i]]
        return l
```
