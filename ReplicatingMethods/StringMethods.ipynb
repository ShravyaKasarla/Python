```python
class MyString:
    
    def lower(x):    
        l = ''
        for i in x:        
            if ord(i) >= 65 and ord(i) <= 90:
                l = l + chr(ord(i)+32)
            else:
                l = l + i 
        return l


    def islower(x):    
        for i in x:  
            if ord(i) >=65 and ord(i) <=90:            
                return False        
        return True


    def upper(x):
        u = ''
        for i in x:        
            if ord(i) >= 97 and ord(i) <= 122:
                u = u + chr(ord(i)-32)
            else:
                u = u + i 
        return u


    def isupper(x):   
        for i in x:  
            if ord(i) >= 97 and ord(i) <= 122:            
                return False        
        return True


    def capitalize(x):    
        new = ''
        for i in range(len(x)):        
            if i == 0:
                if (ord(x[i]) >=97 and ord(x[i]) <=122):
                    new += chr(ord(x[i])-32)
                else:
                    new += x[i]
            elif (ord(x[i]) >=65 and ord(x[i]) <=90):
                new += chr(ord(x[i])+32)
            else:
                new+= x[i]

        return new

    
    def swapcase(x):   
        new=''
        for i in x:
            if (ord(i) >=65 and ord(i) <=90):
                new += chr(ord(i)+32)
            elif (ord(i) >=97 and ord(i) <=122):
                new += chr(ord(i)-32)
            else:
                new += i
        return new
    

    def title(x):    
        new=''
        for i in range(len(x)):
            if i == 0 or (x[i-1] == ' ' or x[i-1] == "'"):
                if (ord(x[i]) >=97 and ord(x[i]) <=122): 
                    new += chr(ord(x[i])-32)
                else:
                    new += x[i]          
            elif (ord(x[i]) >= 65 and ord(x[i]) <=90):
                new += chr(ord(x[i])+32)
            else: 
                new = new + x[i]
        return new   
    
    
    def lstrip(x,sep = ' '):    
        start = 0
        for i in range(len(x)):
            if x[i] in sep: 
                start = i+1
            else:
                break
        return x[start::]     
    
    
    def rstrip(x,sep = ' '):
        stop = len(x)
        for i in range(len(x)):
            if x[-(i+1)] in sep:
                stop = len(x)-1-i
            else:
                break
        return x[:stop:] 
    
    
    def strip(x,sep = ' '):    
        start, stop = 0, len(x)
        for i in range(len(x)):
            if x[i] in sep: 
                start = i+1
            else:
                break
        for i in range(len(x)):
            if x[-(i+1)] in sep:
                stop = len(x)-1-i
            else:
                break
        return x[start:stop:]
    
    
    def removeprefix(x,prefix = ' '):
        if prefix[0] == x[0] and prefix in x:        
            return x[len(prefix)::] 
        return x
    
    
    def removesuffix(x,suffix = ' '):
        if suffix[-1] == x[-1] and suffix in x: 
            return x[:-len(suffix)]  
        return x
    
    
    def split(x,sep =' ',  maxsplit= -1): 
        new=''
        splitnew = []

        for i in range(len(x)):    
            if x[i] != sep:
                new += x[i]
            elif x[i] != ' ' or new != '':
                splitnew = splitnew + list([new])
                new=''
                if len(splitnew) == maxsplit:
                    new = x[i+1:]
                    break
        if x[i] != ' ' or new != '':
            splitnew = splitnew + list([new])
        return splitnew
    
    
    
    def rsplit(x,sep = ' ',  maxsplit= -1):
        new=''
        splitnew = []

        for i in range(len(x)-1, -1 ,-1):    
            if x[i] != sep:
                new += x[i]
            elif x[i] != ' ' or new != '':
                splitnew = splitnew + list([new[::-1]])
                new=''
                if len(splitnew) == maxsplit:
                    new = x[i+1:]
                    break
        if x[i] != ' ' or new != '':
            splitnew = splitnew + list([new])
        return splitnew[::-1]
    
    
    
    def partition(x,sep): 
        if sep in x:
            for i in range(len(x)):
                if x[i] != sep:
                    stop = i+1
                else:
                    break
            return x[:stop], sep, x[stop+1:]    
        else:
            return x, '', ''
        
        
        
    def rpartition(x,sep):
        if sep in x:
            for i in range(len(x)):
                if x[-(i+1)] != sep:
                    stop = -(i+1)-1
                else:
                    break
            return x[:stop], sep, x[stop+1:]     
        else:
            return '', '', x
        

        
    def zfill(x,width):
        if len(x) >= width:
            return x 
        if x[0] in ('+','-'):
            return x[0] + str(0)*(width-len(x)) + x[1:]
        return str(0)*(width-len(x)) + x 
    
    
    
    def ljust(x,width, fillchar = ' '): 
        if len(x) >= width:
            return x 
        fillcount = width - len(x)
        return x + fillchar*fillcount
    
    
    
    def rjust(x,width, fillchar = ' '):
        if len(x) >= width:
            return x 
        fillcount = width - len(x)
        return fillchar*fillcount + x
    
    
    def center(x,width, fillchar = ' '):
        if len(x) >= width:
            return x  
        fillcount = width - len(x)
        if fillcount % 2 == 0:
            return fillchar*int((fillcount/2)) + x + fillchar*int((fillcount/2))
        else:
            return fillchar*int(((fillcount-1)/2)) + x + fillchar*int(((fillcount+1)/2))

    
    def find(x,sub, *args):
        if args == ():
            start, end = 0, len(x) 
        else:
            start, end = args[0], args[1]

        sl = len(sub)
        if sub in x:
            for i in range(start, end):
                if x[i:i+sl] == sub:
                    return i
        return -1
    
    
    def rfind(x,sub, *args):
        if args == ():
            start, end = 0, len(x) 
        else:
            start, end = args[0], args[1]

        sl = len(sub)
        if sub in x:
            for i in range(end, start, -1):
                if x[i-sl:i] == sub:
                    return (i-sl)
        return -1
    
    
    
    def index(x,sub, *args):
        if args == ():
            start, end = 0, len(x) 
        else:
            start, end = args[0], args[1]

        sl = len(sub)
        if sub in x:
            for i in range(start, end):
                if x[i:i+sl] == sub:
                    return i
        raise ValueError(f'substring not found')
        
        
    
    def rindex(x,sub, *args):
        if args == ():
            start, end = 0, len(x) 
        else:
            start, end = args[0], args[1]

        sl = len(sub)
        if sub in x:
            for i in range(end, start, -1):
                if x[i-sl:i] == sub:
                    return (i-sl)
        raise ValueError(f'substring not found') 
        
        
        
    def replace(x,old, new, count = -1):
        ol, nl = len(old), len(new)
        if old in x:
            i, c = 0, 0
            while i <= len(x):
                if x[i:ol+i] == old:
                    x = x[:i]+new+x[i+ol:]
                    i+=nl
                    c +=1
                    if c == -1:
                        continue
                    elif c == count:
                        break
                i+=1
            return x
        return x
    

    
    def count(x,sub, *args):
        if args == ():
            start, end = 0, len(x) 
        else:
            start, end = args[0], args[1]
        c = 0
        sl = len(sub)
        if sub in x and sub != '':
            i = start
            while i <= end:
                if x[i:i+sl] == sub:
                    c+=1
                    i = i+sl
                else:
                    i += 1
            return c   
        elif sub == '':
            return len(x)+1
        else:
            return 0         
        
        
        
    def startswith(x,prefix, *args):
        if args == ():
            start, end = 0, len(x)
        else:
            start, end = args[0], args[1]

        for i in range(start,end):
            for i in prefix:
                if i == x[:len(prefix)]:
                    return True
            return False
        
        
        
    def endswith(x,suffix, *args):
        if args == ():
            start, end = 0, len(x)
        else:
            start, end = args[0], args[1]

        for i in suffix:
            if i == x[len(x)-len(suffix):]:
                return True
        return False
    
    
    
    def isalpha(x): 
        if x != '':
            for i in x:
                if ord(i) not in range(65,91) and ord(i) not in range(97,123):
                    return False
            return True
        return False
    
    
    
    def isalnum(x):
        if x != '':
            for i in x:
                if ord(i) not in range(65,91) and ord(i) not in range(97,123) and ord(i) not in range(48,58):
                    return False
            return True
        return False
    
    
    
    def isascii(x):
        if x != '':
            for i in x:
                if ord(i) not in range(0,128):
                    return False
        return True
    
 

    def isnumeric(x):
        if x != '':
            for i in x:
                if ord(i) not in range(48,58):
                    return False
        return True
    
    
    
    def isprintable(x): 
        for i in x:
            if ord(i) not in range(32,256):
                return False
        return True
    
    
    
    def isspace(x):  
        if x != '':
            for i in x:
                if ord(i) != 32:
                    return False
            return True
        return False
    
    
    
    def istitle(x):
        for i in range(len(x)):
            if i == 0 or (x[i-1] == ' ' or x[i-1] == "'") :
                if ord(x[i]) not in range(65, 91):
                    return False
            elif ord(x[i]) in range(65, 91):
                return False
        return True   
    
    
    
    def join(itr, sep=''):
        new = ''
        for i in itr:
            new += (i + sep)
        return new[:len(new)-1]
    
    
    
    import multipledispatch
    from multipledispatch import dispatch
    @dispatch(str,str)
    def concat(self,a,b):
        return a+b

    @dispatch(str,int)
    def repeat(self,a,b):
        return a*b
    
    @dispatch(str)
    def length(self,a):
        count=0
        for i in a:
            count+=1
        return count
```
