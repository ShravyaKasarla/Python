```python
class MyList:
    def __init__(self,lst):
        self.lst = lst
        
    #1.
    def append(self, lst, add_value):
        '''Append object to the end of the list.'''
        lst[len(lst):len(lst)] = [add_value] 
      
    #2.
    def clear(self, lst):
        '''Remove all items from list'''
        del lst[:]
    
    #3.
    def copy(self, lst):
        '''Return a shallow copy of the list.'''
        return lst[:]
    
    #4.
    def count(self, lst, value):
        '''Remove all items from list.'''          
        if value in lst:
            c = 0
            for i in lst:
                if i == value:
                    c += 1
            return c
        return 0  
    
    #5.
    def extend(self, lst, add_value):
        '''Extend list by appending elements from the iterable.'''
        lst[len(lst):len(lst)] = add_value   
        
    #6.    
    def index(self, lst, value, *args):
        '''Return first index of value.
        Raises ValueError if the value is not present.'''
        if args == ():
            start, end = 0, len(lst) 
        elif len(args) == 1:
            start = args[0]
            end = len(lst)
        elif len(args) == 2:
            start, end = args[0], args[1]
        else:
            raise TypeError(f'index expected at most 3 arguments, got 4')
        for i in range(start,end):
            if lst[i] == value:
                return i
        raise ValueError(f'{value} is not in list')  
        
    #7.
    def insert(self, lst, index, value):
        ''' Insert object before index.'''
        lst[index:index] = list([value])
        
    #8.
    def pop(self, lst, index = -1):
        '''Remove and return item at index (default last).
        Raises IndexError if list is empty or index is out of range.'''
        pop_value = lst[index]
        if index != -1:
            lst[:] = lst[:index] + lst[index+1:]
        else:
            lst[:] = lst[:index]
        return pop_value
    
    #9.
    def remove(self, lst, value):
        if value in lst:    
            for i in lst:   
                if i == value:
                    lst[:] = lst[:i-1] + lst[i:]
        else:        
            raise ValueError('remove(lst, x): x not in list')
          
    #10.
    def reverse(self, lst):
        '''Remove first occurrence of value.
        Raises ValueError if the value is not present.'''
        lst[:] = lst[::-1]
        
    #11.
    def sort(self, lst, reverse = False):
        '''Sort the list in ascending order and return None.'''
        n = 0
        for ele in lst:
            n += 1
        #bubblesort    
        for i in range(n):
            swapped = False
            for j in range(n-i-1):
                if (lst[j]>lst[j+1]):
                    lst[j], lst[j+1] = lst[j+1], lst[j]
                    swapped = True
            if swapped == False:
                break
        if reverse == True:
            lst[:] = lst[::-1]
    
```

```python
class MyTuple(MyList):
    def __init__(self,tup):
        self.tup = tup
    
    def tupleCount(self, tup, value):
        '''Return number of occurrences of value.'''
        return self.count(tup,value)
    
    
    def tupleIndex(self, tup, value, *args):
        '''Return first index of value.
        Raises ValueError if the value is not present.'''
        return self.index(tup, value, *args)
```
